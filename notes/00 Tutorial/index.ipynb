## 一、需要的软硬件资源：算力和开发

### 1、为什么使用 NumPy？
- 优化的数据操作
- 矩阵计算

### 2、什么是 CUDA？
- GPU 计算
- 针对 N 卡

### 3、为什么使用 pyTorch？
- GPU 计算
- 无缝切换 CPU 和 GPT
- 机器学习专用库
- 多卡部署
- 主流支持：N 卡等

### 4、为什么使用 HuggingFace？
- 兼容 pyTorch 和 Tensorflow
- 云端发布和使用模型

### 5、各云平台机器学习提供了什么？
- 一站式训练资源：数据、模型、算力租用、发布

### 6、为什么使用langchain？
- 大模型开发

### 7、GPU选型
- 4090是好的起点（2-5万）
- A100等可做实际部署（5-50万）
