## 绘制散点图：

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.scatter(x, y)
plt.show()
```

## 绘制柱状图：

```python
import matplotlib.pyplot as plt

x = ['A', 'B', 'C', 'D', 'E']
y = [3, 10, 7, 5, 3]

plt.bar(x, y)
plt.show()
```

## 绘制饼图：

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```
